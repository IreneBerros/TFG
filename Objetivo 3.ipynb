### Objetivo 3:
Caida campo magn√©tico

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#radios
a=6371.2e3 #tierra (m)
b=1212e3 #nucleo interno (m)
c=3485e3 #nucleo externo (m)

#velocidades
V_menos=0.5e-3
V=0.75e-3
V_mas=1e-3


#velocidad constante
def r_fun(v,t):
    return b+(v*t)

#velocidad radio
def rr_fun(v,t):
    return b*((1+(2*v*t/b))**(1/2))

#velocidad area
def ra_fun(v,t):
    return b*((1+(3*v*t/b))**(1/3))

#velocidad volumen 
def rv_fun(v,t):
    return b*((1+(4*v*t/b))**(1/4))

#tiempos
t=np.linspace(0,5e10,1000)



#campo magn√©tico
def B_fun(r):
    return 60000*((c**3-r**3)/(c**3-b**3))


#lineal
B_l=B_fun(r_fun(V,t))
Bmen_l=B_fun(r_fun(V_menos,t))
Bmas_l=B_fun(r_fun(V_mas,t))

#radio
B_r=B_fun(rr_fun(V,t))
Bmen_r=B_fun(rr_fun(V_menos,t))
Bmas_r=B_fun(rr_fun(V_mas,t))

#area
B_a=B_fun(ra_fun(V,t))
Bmen_a=B_fun(ra_fun(V_menos,t))
Bmas_a=B_fun(ra_fun(V_mas,t))

#volumen
B_v=B_fun(rv_fun(V,t))
Bmen_v=B_fun(rv_fun(V_menos,t))
Bmas_v=B_fun(rv_fun(V_mas,t))


#graficos
fig, [[ax33, ax34], [ax43,ax44]]  = plt.subplots(nrows=2, ncols=2, figsize=(10,6), constrained_layout=True)

fig.suptitle('Ca√≠da del Campo Magn√©tico',fontsize=15, weight='bold')

#velocidad linal

ax33.plot(t,B_l,label='V',color='steelblue')
ax33.plot(t,Bmen_l,label='V +- 0.25',linestyle='--',color='lightskyblue')
ax33.plot(t,Bmas_l,linestyle='--',color='lightskyblue')
ax33.set_ylim(0,62000)
ax33.grid()
ax33.legend(loc=1)
ax33.set_title('Lineal')
ax33.set_ylabel("CM (nT)")
ax33.set_xlabel("Tiempo (a√±os)")

#velocidad radio

ax34.plot(t,B_r,label='V',color='crimson');
ax34.plot(t,Bmen_r,label='V +- 0.25',linestyle='--',color='lightpink')
ax34.plot(t,Bmas_r,linestyle='--',color='lightpink')
ax34.set_ylim(0,62000)
ax34.grid()
ax34.legend(loc=1)
ax34.set_title('Dependencia unidimensional')
ax34.set_ylabel("CM (nT)")
ax34.set_xlabel("Tiempo (a√±os)")

#velocidad area

ax43.plot(t,B_a,label='V',color='chocolate');
ax43.plot(t,Bmen_a,label='V +- 0.25',linestyle='--',color='orange')
ax43.plot(t,Bmas_a,linestyle='--',color='orange')
ax43.set_ylim(0,62000)
ax43.grid()
ax43.legend(loc=1)
ax43.set_title('Dependencia bidimensional')
ax43.set_ylabel("CM (nT)")
ax43.set_xlabel("Tiempo (a√±os)")

#velocidad volumen

ax44.plot(t,B_v,label='V',color='darkgreen');
ax44.plot(t,Bmen_v,label='V +- 0.25',linestyle='--',color='mediumseagreen')
ax44.plot(t,Bmas_v,linestyle='--',color='mediumseagreen')
ax44.set_ylim(0,62000)
ax44.grid()
ax44.legend(loc=1)
ax44.set_title('Dependencia tridimensional')
ax44.set_ylabel("CM (nT)")
ax44.set_xlabel("Tiempo (a√±os)");

plt.savefig("campo.jpg")
```
