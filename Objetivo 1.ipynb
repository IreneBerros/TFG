### Objetivo 1: 
Aumento de radio del nucleo interno

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#radios
a=6371.2e3 #tierra (m)
b=1212e3 #nucleo interno (m)
c=3485e3 #nucleo externo (m)

#velocidades
V_menos=0.5e-3
V=0.75e-3
V_mas=1e-3


#velocidad constante
def r_fun(v,t):
    return b+(v*t)

#velocidad radio
def rr_fun(v,t):
    return b*((1+(2*v*t/b))**(1/2))

#velocidad area
def ra_fun(v,t):
    return b*((1+(3*v*t/b))**(1/3))

#velocidad volumen 
def rv_fun(v,t):
    return b*((1+(4*v*t/b))**(1/4))

#tiempos
t=np.linspace(0,5e10,1000)


#graficos
fig, [[ax11, ax12], [ax21,ax22]]  = plt.subplots(nrows=2, ncols=2, figsize=(10,6), constrained_layout=True)

fig.suptitle("Aumento del radio",fontsize=15, weight='bold')

#velocidad lineal
ax11.plot(t,r_fun(V,t),label='V',color='steelblue');
ax11.plot(t,r_fun(V_menos,t),label='V +- 0.25',linestyle='--',color='lightskyblue')
ax11.plot(t,r_fun(V_mas,t),linestyle='--',color='lightskyblue')
ax11.set_ylim(1e6,c)
ax11.grid()
ax11.legend(loc=4)
ax11.set_title('Lineal')
ax11.set_ylabel("Radio (m)")
ax11.set_xlabel("Tiempo (aos)")

#velocidad radio
ax12.plot(t,rr_fun(V,t),label='V',color='crimson');
ax12.plot(t,rr_fun(V_menos,t),label='V +- 0.25',linestyle='--',color='lightpink')
ax12.plot(t,rr_fun(V_mas,t),linestyle='--',color='lightpink')
ax12.set_ylim(1e6,c)
ax12.grid()
ax12.legend(loc=4)
ax12.set_title('Dependencia unidimensional')
ax12.set_ylabel("Radio (m)")
ax12.set_xlabel("Tiempo (aos)")

#velocidad area
ax21.plot(t,ra_fun(V,t),label='V',color='chocolate');
ax21.plot(t,ra_fun(V_menos,t),label='V +- 0.25',linestyle='--',color='orange')
ax21.plot(t,ra_fun(V_mas,t),linestyle='--',color='orange')
ax21.set_ylim(1e6,c)
ax21.grid()
ax21.legend(loc=4)
ax21.set_title('Dependencia bidimensional')
ax21.set_ylabel("Radio (m)")
ax21.set_xlabel("Tiempo (aos)")

#velocidad volumen
ax22.plot(t,rv_fun(V,t),label='V',color='darkgreen');
ax22.plot(t,rv_fun(V_menos,t),label='V +- 0.25',linestyle='--',color='mediumseagreen')
ax22.plot(t,rv_fun(V_mas,t),linestyle='--',color='mediumseagreen')
ax22.set_ylim(1e6,c)
ax22.grid()
ax22.legend(loc=4)
ax22.set_title('Dependencia tridimensional')
ax22.set_ylabel("Radio (m)")
ax22.set_xlabel("Tiempo (aos)");

plt.savefig("radio.jpg")
```

### Tiempo en llegar a c (radio nucleo externo)

```python
#v linal
t=1
i=0
while i<(c-1000):
    i=r_fun(V,t)
    t=t+10e5

print('Con velocidad constante tarda aprox:' ,t, 'aos')
```

```python
#v radio
t=1
i=0
while i<(c-1000):
    i=rr_fun(V,t)
    t=t+10e5

print('Con velocidad constante tarda aprox:' ,t, 'aos')
```

```python
#v dependiente del rea
t=1
i=0
while i<(c-100):
    i=ra_fun(V,t)
    t=t+10e5

print('Con velocidad dependiente del rea tarda aprox:' ,t, 'aos')
```

```python
#v dependiente del volumen
t=1
i=0
while i<(c-1000):
    i=rv_fun(V,t)
    t=t+10e5

print('Con velocidad dependiente del volumen tarda aprox:' ,t, 'aos')
```
