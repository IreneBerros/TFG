### Objetivo 2:
Estimaci√≥n origen del nucleo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#radios
a=6371.2e3 #tierra (m)
b=1212e3 #nucleo interno (m)
c=3485e3 #nucleo externo (m)

#velocidades
V_menos=0.5e-3
V=0.75e-3
V_mas=1e-3


#velocidad constante
def r_fun(v,t):
    return b+(v*t)

#velocidad radio
def rr_fun(v,t):
    return b*((1+(2*v*t/b))**(1/2))

#velocidad area
def ra_fun(v,t):
    return b*((1+(3*v*t/b))**(1/3))

#velocidad volumen 
def rv_fun(v,t):
    return b*((1+(4*v*t/b))**(1/4))



t=np.linspace(0,-2.5e9,1000)


#velocidad lineal

r_l=[]
tl=0
t_l=[]
l=b
while l>=0:
    l=r_fun(V,tl)
    tl=tl-1e7
    t_l.append(tl)
    r_l.append(l)
    
    
r_lVmas=[]
tl=0
t_lVmas=[]
l=b
while l>=0:
    l=r_fun(V_mas,tl)
    tl=tl-1e7
    t_lVmas.append(tl)
    r_lVmas.append(l)
    
    
r_lVmen=[]
tl=0
t_lVmen=[]
l=b
while l>=0:
    l=r_fun(V_menos,tl)
    tl=tl-1e7
    t_lVmen.append(tl)
    r_lVmen.append(l)
    
    


#velocidad radio

r_r=[]
tr=0
t_r=[]
r=b
while r>=10:
    r=rr_fun(V,tr)
    tr=tr-1e6
    t_r.append(tr)
    r_r.append(r)


    
r_rVmas=[]
tr=0
t_rVmas=[]
r=b
while r>=10:
    r=rr_fun(V_mas,tr)
    tr=tr-1e6
    t_rVmas.append(tr)
    r_rVmas.append(r)



r_rVmen=[]
tr=0
t_rVmen=[]
r=b
while r>=10:
    r=rr_fun(V_menos,tr)
    tr=tr-1e6
    t_rVmen.append(tr)
    r_rVmen.append(r)

    

    
#velocidad area

r_a=[]
t_a=[]
ta=0
a=b

while a>=1e4:
    a=ra_fun(V,ta)
    ta=ta-1e2
    t_a.append(ta)
    r_a.append(a)
    
r_aVmas=[]
t_aVmas=[]
ta=0
a=b

while a>=1e4:
    a=ra_fun(V_mas,ta)
    ta=ta-1e2
    t_aVmas.append(ta)
    r_aVmas.append(a)
    

r_aVmen=[]
t_aVmen=[]
ta=0
a=b

while a>=1e4:
    a=ra_fun(V_menos,ta)
    ta=ta-1e2
    t_aVmen.append(ta)
    r_aVmen.append(a)
    
    
    
#velocidad volumen

r_v=[]
t_v=[]
tv=0
v=b

while v>=1e4:
    v=rv_fun(V,tv)
    tv=tv-1e3
    t_v.append(tv)
    r_v.append(v)

    
r_vVmas=[]
t_vVmas=[]
tv=0
v=b

while v>=1e4:
    v=rv_fun(V_mas,tv)
    tv=tv-1e3
    t_vVmas.append(tv)
    r_vVmas.append(v)
    

r_vVmen=[]
t_vVmen=[]
tv=0
v=b

while v>=1e4:
    v=rv_fun(V_menos,tv)
    tv=tv-1e3
    t_vVmen.append(tv)
    r_vVmen.append(v)
    
    
    
        
#graficos
fig, [[ax55, ax56], [ax65,ax66]]  = plt.subplots(nrows=2, ncols=2, figsize=(10,6), constrained_layout=True)

fig.suptitle('Comienzo de solidificaci√≥n del n√∫cleo',fontsize=15, weight='bold')

y = np.arange(0, 1400e3, 1)

#velocidad lineal

ax55.plot(t_l,r_l,label='V',color='steelblue');
ax55.plot(t_lVmas,r_lVmas,label='V +- 0.25',linestyle='--',color='lightskyblue')
ax55.plot(t_lVmen,r_lVmen,linestyle='--',color='lightskyblue')
ax55.set_ylim(0,1400e3)
ax55.set_xlim(-2.5e9,0)
ax55.grid()
ax55.legend(loc=2)
ax55.set_title('Lineal')
ax55.set_ylabel("Radio (m)")
ax55.set_xlabel("Tiempo (a√±os)")

#velocidad radio

ax56.plot(t_r,r_r,label='V',color='crimson');
ax56.plot(t_rVmas,r_rVmas,label='V +- 0.25',linestyle='--',color='lightpink')
ax56.plot(t_rVmen,r_rVmen,linestyle='--',color='lightpink')
ax56.set_ylim(0,1400e3)
ax56.set_xlim(-2.5e9,0)
ax56.grid()
ax56.legend(loc=2)
ax56.set_title('Dependencia unidimensional')
ax56.set_ylabel("Radio (m)")
ax56.set_xlabel("Tiempo (a√±os)")

#velocidad area

ax65.plot(t_a,r_a,label='V',color='chocolate')
ax65.plot(t_aVmas,r_aVmas,label='V +- 0.25',linestyle='--',color='orange')
ax65.plot(t_aVmen,r_aVmen,linestyle='--',color='orange')
ax65.set_ylim(0,1400e3)
ax65.set_xlim(-2.5e9,0)
ax65.grid()
ax65.legend(loc=2)
ax65.set_title('Dependencia bidimensional')
ax65.set_ylabel("Radio (m)")
ax65.set_xlabel("Tiempo (a√±os)")

#velocidad volumen

ax66.plot(t_v,r_v,label='V',color='darkgreen');
ax66.plot(t_vVmas,r_vVmas,label='V +- 0.25',linestyle='--',color='mediumseagreen')
ax66.plot(t_vVmen,r_vVmen,linestyle='--',color='mediumseagreen')
ax66.set_ylim(0,1400e3)
ax66.set_xlim(-2.5e9,0)
ax66.grid()
ax66.legend(loc=2)
ax66.set_title('Dependencia tridimensional')
ax66.set_ylabel("Radio (m)")
ax66.set_xlabel("Tiempo (a√±os)");


plt.savefig("Radio_cero.jpg")
```
